```python
# the following code writes the runfile for EcoSIM
#relationship between weather file year and optfile year, they don't have to be equal, optfile will overwride
#grid cooridinates, left upper corner, lower right corner
#[west,north,east,south]
grid_coordinates=[1,1,1,1]
#name of the site file, which defines site information
site_file='site.example'
#topography file
top_file='top.example'

#
#ecosim simulation is defined by scenarios, each scenario has segments of scenes,   
#number of scenes in a segment of a scenario, each element represents a scenario, NAX
#num_scenes->NAX
num_scenes=1
#number of segments of a scenario, where each segment repeats the scenes
#num_runcycles->NDX, ususally set to 1
num_runcycles=1
#num_scenes and num_runcycles should be have the same number of columns
#each scene is made up of multiple periods, and each period has multiple years
#NAY->NA(NEX), NEX=1,NAX
spinup_len=5
spinup_clim_year0=1980
year0=1950
regrun_len=10
regrun_cyc=1
ryear0=1960
#NDY->ND(NEX)
spinup_cyc=6

year_config={'weather_file':'US-FufwYYYY','solver_config':'optYYYY','land_mgmt':'NO','pft_mgmt':'pft1',\
            'hr_C_vars':'NO','hr_h2o_vars':'NO','hr_N_vars':'NO','hr_P_vars':'NO','hr_heat_vars':'NO',\
             'day_C_vars':'NO','day_h2o_vars':'NO','day_N_vars':'NO','day_P_vars':'NO','day_heat_vars':'NO'}

print('%d %d %d %d'%(grid_coordinates[0],grid_coordinates[1],grid_coordinates[2],grid_coordinates[3]))
print('%s'%site_file)
print('%s'%top_file)
print('%d %d'%(num_scenes,num_runcycles))
#scene 1, spinup
print('%d %d'%(spinup_len,spinup_cyc))
for k in range(spinup_len):
    print(year_config['weather_file'].replace("YYYY","%4d"%(spinup_clim_year0+k)))
    print(year_config['solver_config'].replace("YYYY","%4d"%(year0+k)))
    print(year_config['land_mgmt'].replace("NO","mylu"))
    print(year_config['pft_mgmt'].replace("pft1","mypft"))
    print(year_config['hr_C_vars'].replace("NO","NO"))
    print(year_config['hr_h2o_vars'].replace("NO","NO"))
    print(year_config['hr_N_vars'].replace("NO","NO"))
    print(year_config['hr_P_vars'].replace("NO","NO"))
    print(year_config['hr_heat_vars'].replace("NO","NO"))
    print(year_config['day_C_vars'].replace("NO","dc"))
    print(year_config['day_h2o_vars'].replace("NO",'dw'))
    print(year_config['day_N_vars'].replace("NO","NO"))
    print(year_config['day_P_vars'].replace("NO","NO"))
    print(year_config['day_heat_vars'].replace("NO","dh"))
#scene 2,regular run

for k in range(spinup_len):
    print(year_config['weather_file'].replace("YYYY","%4d"%(spinup_clim_year0+k)))
    print(year_config['solver_config'].replace("YYYY","%4d"%(year0+k)))
    print(year_config['land_mgmt'].replace("NO","mylu"))
    print(year_config['pft_mgmt'].replace("pft1","mypft"))
    print(year_config['hr_C_vars'].replace("NO","NO"))
    print(year_config['hr_h2o_vars'].replace("NO","NO"))
    print(year_config['hr_N_vars'].replace("NO","NO"))
    print(year_config['hr_P_vars'].replace("NO","NO"))
    print(year_config['hr_heat_vars'].replace("NO","NO"))
    print(year_config['day_C_vars'].replace("NO","dc"))
    print(year_config['day_h2o_vars'].replace("NO",'dw'))
    print(year_config['day_N_vars'].replace("NO","NO"))
    print(year_config['day_P_vars'].replace("NO","NO"))
    print(year_config['day_heat_vars'].replace("NO","dh"))
#scene 2,regular run
print('%d %d'%(regrun_len,regrun_cyc))
for k in range(regrun_len):
    print(year_config['weather_file'].replace("YYYY","%4d"%(ryear0+k)))
    print(year_config['solver_config'].replace("YYYY","%4d"%(ryear0+k)))
    print(year_config['land_mgmt'].replace("NO","mylu"))
    print(year_config['pft_mgmt'].replace("pft1","mypft"))
    print(year_config['hr_C_vars'].replace("NO","NO"))
    print(year_config['hr_h2o_vars'].replace("NO","NO"))
    print(year_config['hr_N_vars'].replace("NO","NO"))
    print(year_config['hr_P_vars'].replace("NO","NO"))
    print(year_config['hr_heat_vars'].replace("NO","NO"))
    print(year_config['day_C_vars'].replace("NO","dc"))
    print(year_config['day_h2o_vars'].replace("NO",'dw'))
    print(year_config['day_N_vars'].replace("NO","NO"))
    print(year_config['day_P_vars'].replace("NO","NO"))
    print(year_config['day_heat_vars'].replace("NO","dh"))    

print('%d %d'%(0,0))








    
        
```
